```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers...

```python
# convert data type to float in "Average_price" column
prop_tall['Average_price'] = prop_tall['Average_price'].astype(np.float64)
```

#### 2.5. Cleaning the data (part 3)

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
# check how many Boroughs exist in the dataframe
prop_tall.drop_duplicates(subset='London_Borough') \
        .sort_values('London_Borough').reset_index(drop=True)
```

```python
# chekc the data with the Borough named by 'Unnamed:...'

prop_tall[prop_tall['London_Borough'].str.contains('Unnamed:')]
```

```python
# remove all data for 'Unnamed:...' Boroughs
prop_clean = prop_tall[~prop_tall['London_Borough'].str.contains('Unnamed:')]

# check again the dataset with unique Boroughs
prop_clean.drop_duplicates(subset='London_Borough') \
                .sort_values('London_Borough').reset_index(drop=True)
```

```python
""" Remove the list of non-Boroughs. 
    Note that 'City of London' is the 33rd principal division of Greater London 
    but it is not a London borough according to wikipedia 
    (https://en.wikipedia.org/wiki/List_of_London_boroughs)
    After this process, the number of Boroughs should be 32!!
    """

nonBoroughs = ['City of London', 'EAST MIDLANDS', 'EAST OF ENGLAND', 'England', 'Inner London', \
               'LONDON', 'NORTH EAST', 'NORTH WEST', 'Outer London',\
               'SOUTH EAST', 'SOUTH WEST', 'WEST MIDLANDS', 'YORKS & THE HUMBER']

prop_df = prop_clean[~prop_clean['London_Borough'].isin(nonBoroughs)]

#prop_clean['London_Borough'].unique()

prop_df.drop_duplicates(subset='London_Borough') \
                .sort_values('London_Borough').reset_index(drop=True)
```

```python
""" simple check which Borough has higher averge housing price """

prop_df.drop_duplicates(subset='London_Borough') \
                .sort_values('Average_price', ascending=False).reset_index(drop=True)
```

```python
# check if there is any missing values or null data
prop_df.isna().any()
```

```python
# check the DataFrame information finally
print(prop_df.info())

print("dimension of the dataFrame: ", prop_df.shape)
```

#### 2.6. Visualizing the data

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
# sorted list containing unique Borough names
unique_Boroughs = sorted(prop_df['London_Borough'].unique())

print(unique_Boroughs)
```

```python
''' Draw a linear plot '''

fig, ax = plt.subplots()

# draw line plots of average price as a function of Month for each Borough
for iBorough in unique_Boroughs:
    prices = prop_df[prop_df['London_Borough']==iBorough]
    prices.plot(kind='line', x='Month', y='Average_price', ax=ax, legend=True, label=iBorough)

ax.set_ylabel('Average price')
ax.legend(loc='upper left', fontsize=6., ncol=2)

# the tick format of y-axis is converted to 'M' unit that indicates million (10^6)
ax.yaxis.set_major_formatter(lambda y, pos: '%.1f M'%(y/1e6))
```

### 3. Modeling

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
''' function that calculate Average price for 2018 / Average price for 1998 
    Note that I inversed the ratio (see the guided equation was Price_1998 / Price_2018)
    to estimate which ratio the price would be increased between two years. '''

def create_price_ratio(df):
    pr_1998 = df.loc[1998,'Average_price']
    pr_2018 = df.loc[2018,'Average_price']
    
    return pr_2018 / pr_1998
```

```python
# initiate the list that will contain the ratio data (Price_2018/Price_1998) for each Borough
ratio = []

# compute the ratio for each Borough 
for iBorough in unique_Boroughs:
    ratio_Bor = create_price_ratio(prop_df_grbyYear.loc[iBorough])
    ratio.append(ratio_Bor)

#check the list of data
print(unique_Boroughs)
print(ratio)
```

```python
''' create the dictionary-of-list that contains the calculated ratio above, 
    and then convert it to pandas DataFrame '''

# dictionary of list
dol = {'London_Borough':unique_Boroughs, 'ratios':ratio}

# convert the dictionary of list to pandas DataFrame
df_ratios = pd.DataFrame(dol).set_index('London_Borough')
df_ratios
```

```python
# check the sorted DataFrame to see which Borough has higher ratio (i.e., the price is more increased)
df_ratios.sort_values("ratios", ascending=False)
```

```python
# tag the top10 Borough with high ratio to remark the different color belowe 
df_ratio_top10 = df_ratios.sort_values("ratios", ascending=False).head(10)
df_ratio_top10.index
```

```python
''' Draw Bar plot for the ratio between Price_2018 and Price_1998 '''

fig, ax = plt.subplots()

# color will be assigned to 'red' for top10 high increased Borough and
# 'blue' for the rest
c = []
for iBorough in unique_Boroughs:
    c.append('red') if iBorough in df_ratio_top10.index else c.append('blue')

df_ratios.plot(kind='bar', y='ratios', ax=ax, legend=False, color=c)

ax.set_ylabel(r'price$_{2018}$ / price$_{1998}$')
ax.tick_params(axis='x', labelsize=11)
```
