# Joining

```python
d1 = pd.DataFrame({'id': [1, 2, 10, 12], 
     'val1': ['a', 'b', 'c', 'd']})
d1
```

```python
d2 = pd.DataFrame({'id': [1, 2, 9, 8],
     'val1': ['p', 'q', 'r', 's']})
d2
```

## inner join

```python
d3 = pd.merge(d1,d2, on='id', how='inner', suffixes=('_left','_right'))
d3
```

```python
d3 = d1.merge(d2, on='id', how='inner', suffixes=('_left','_right'))
d3
```

```python
d3 = pd.merge(d1,d2, left_on='id', right_on='id', how='inner', suffixes=('_left','_right'))
d3
```

## full outer join

```python
d4 = pd.merge(d1,d2, left_on='id', right_on='id', how='outer', suffixes=('_left','_right'), indicator=True)
d4
```

```python
''' outer join without intersection '''

d4 = pd.merge(d1,d2, left_on='id', right_on='id', how='outer', suffixes=('_left','_right'))

ind_intersect = (d4['val1_left'].isnull() | d4['val1_right'].isnull())
d5 = d4[ind_intersect]
d5
```

## left outer join

```python
d6 = pd.merge(d1,d2, left_on='id', right_on='id', how='left', suffixes=('_left','_right'), indicator=True)
d6
```

# Concatenation

```python
d_con = pd.concat([d1,d2], join='inner', sort=True, ignore_index=True, axis=0)

d_con
```
